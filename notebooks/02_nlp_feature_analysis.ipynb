# NLP feature analysis
Sentiment, topic, and uncertainty distributions per call.

```python
import pandas as pd
from pathlib import Path

feat = Path("data/processed/features")
for name in ["sentiment_per_call.parquet", "uncertainty_per_call.parquet"]:
    p = feat / name
    if p.exists():
        df = pd.read_parquet(p)
        print(name, df.shape)
        display(df.describe())
```
